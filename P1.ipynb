## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=10):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    
    improved_draw_lines_flag = True
    
    if not improved_draw_lines_flag:
        for line in lines:
            for x1,y1,x2,y2 in line:
                cv2.line(img, (x1, y1), (x2, y2), color, thickness)
    else:
        slope_left = []
        slope_right = []
        x_left = []
        y_left = []
        x_right = []
        y_right = []
        slope_threshold = 0.5

        for line in lines:
            if len(line) != 1 or len(line[0]) != 4:
                continue

            # Unpack the xy value from each point
            for x1,y1,x2,y2 in line:

                # Calculate the slope
                dx = x2 - x1
                dy = y2 - y1
                slope = dy/dx

                # Positive slope is for the left lane line
                if slope > slope_threshold and slope < 1.5:
                    slope_left += [slope]
                    x_left.extend((x1, x2))
                    y_left.extend((y1, y2))

                # Negative slope is for the right lane line
                if slope < -slope_threshold and slope > -1.5:
                    slope_right += [slope]
                    x_right.extend((x1, x2))
                    y_right.extend((y1, y2))

        # Make sure lines were found in frame
        if len(slope_left) == 0 or len(slope_right) == 0:
            return

        # Calculate average position for each lane line
        avg_slope_left = np.mean(slope_left)
        avg_slope_right = np.mean(slope_right)
        avg_x_left = np.mean(x_left)
        avg_y_left = np.mean(y_left)
        avg_x_right = np.mean(x_right)
        avg_y_right = np.mean(y_right)

        # Find intercept for each average lane line (b = y - mx)
        b_left = avg_y_left - avg_slope_left * avg_x_left
        b_right = avg_y_right - avg_slope_right * avg_x_right

        # Corresponds to very bottome of image
        max_y = image.shape[0]
        #max_y = max(max(y_left), max(y_right))
        min_y = min(min(y_left), min(y_right))

        # Find top and bottom x points for both lanes (x = y-b/m)
        left_x_top = int((min_y - b_left)/avg_slope_left)
        left_x_bottom = int((max_y - b_left)/avg_slope_left)
        right_x_top = int((min_y - b_right)/avg_slope_right)
        right_x_bottom = int((max_y - b_right)/avg_slope_right)
        
        cv2.line(img, (left_x_top, min_y), (left_x_bottom, max_y), color, thickness)
        cv2.line(img, (right_x_top, min_y), (right_x_bottom, max_y), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., γ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + γ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, γ)
````

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images_output directory.

#####PIPELINE######
# Read RGB image
# Convert to HLS color space
# Mask every color except white and yellow
# Convert back to RGB color space

## Global Variables ##

# Image directory
path = "test_images/"

# Output directory
output_path = "test_images_output/"

# Set to true to only run 1 image
DEBUG = True

## PARAMETERS ##

# Gaussian Blur Kernel Size
KERNEL_SIZE = 5

# Canny Edge Tresholds
CANNY_HI_THRESH = 150
CANNY_LO_THRESH = 25

# Hough Transform Parameters

RHO = 1
THETA = np.pi/180
H_THRESHOLD = 15
MIN_LINE_LENGTH = 10
MAX_LINE_GAP = 20

#############################################################################
# Function: rgb_hls_color_mask
#
# Description: This function is used to take an RGB
# image and convert it to HSL then filter out the white
# and yellow pixels to apply mask to original image
#
# Parameters:
#            rgb_image: RGB image
#            out_img: RGB image with applied color mask
#
##############################################################################
def rgb_hls_color_mask(rgb_img):
    
    # Convert to HLS color space and mask yellow and white 
    hls_img = cv2.cvtColor(rgb_img, cv2.COLOR_RGB2HLS)
    # Find white pixels on HLS image
    dark = np.array([0, 200, 0], dtype='uint8')
    light = np.array([255, 255, 255], dtype='uint8')
    # Create mask for white in HLS image
    white_mask = cv2.inRange(hls_img, dark, light)

    # Find yellow pixels on HLS image
    dark = np.array([20, 120, 80], dtype='uint8')
    light = np.array([45, 200, 255], dtype='uint8')          
    # Create mask for yellow in hls image
    yellow_mask = cv2.inRange(hls_img, dark, light)

    # Bitwise and both masks with HLS image
    img_masked = cv2.bitwise_and(hls_img, hls_img, mask=(white_mask|yellow_mask)).astype(np.uint8)
    
    # Convert image back to RGB
    img_out = cv2.cvtColor(img_masked, cv2.COLOR_HLS2RGB)

    return img_out

# Loop through each image in test directory
for img_file in os.listdir(path):
    
    # Build image path
    img_path = path + img_file 
    print(img_path)
    
    # Read image with matplot lib
    image = mpimg.imread(img_path)
    
    # Convert image to HLS color space and apply yellow and white color masks
    # image is returned back in RGB color space
    img_out = rgb_hls_color_mask(image)
    
    # Read image with opencv
    #image = cv2.imread(img_path)
    
    # Convert image to grayscale using helper function
    gray_image = grayscale(img_out)
    
    # Create new image name for grayscale image
    gray_img_name = img_file.split('.')[0] + '_grayscale' + img_file.split('.')[1]
    
    # Use gaussian_blur helper to smooth image
    blur_gray_img = gaussian_blur(gray_image, KERNEL_SIZE)

    
#     plt.figure()
#     plt.plot(histr)
#     plt.title('Image Histogram')
#     plt.show()
    
    # Determine threshold paramters using OTSU automatic thresholding
#     CANNY_HI_THRESH, img_thresh = cv2.threshold(blur_gray_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
#     CANNY_LO_THRESH = 0.5 * CANNY_HI_THRESH

    
    # Use canny helper function for canny edge detection
    canny_edge_img = canny(blur_gray_img, CANNY_LO_THRESH, CANNY_HI_THRESH)
    
    # Now use region of interest helper to mask a region of the image
   
    vertices = np.array([[(0,image.shape[0]),(450, 320), (550, 320), (image.shape[1],image.shape[0])]], dtype=np.int32)
    masked_edge_img = region_of_interest(canny_edge_img, vertices)
    
#     cv2.line(image, (0, image.shape[0]), (400, 320), [255, 0, 0], 10)
#     cv2.line(image, (550, 320), (400, 320), [255, 0, 0], 10)
#     cv2.line(image, (550, 320), (image.shape[1],image.shape[0]), [255, 0, 0], 10)
#     plt.imshow(image)
#     plt.show()
    
    # Create copy of image to draw lane lines on
    final_image_lines = np.copy(image)
    
    
    
    # Perform Hough Transform
    #lines = cv2.HoughLinesP(masked_edge_img, RHO, THETA, H_THRESHOLD, np.array([]), minLineLength=MIN_LINE_LENGTH, maxLineGap=MAX_LINE_GAP)
    lines = hough_lines(masked_edge_img, RHO, THETA, H_THRESHOLD, MIN_LINE_LENGTH, MAX_LINE_GAP)
    result = cv2.addWeighted(final_image_lines, 0.8, lines, 1.0, 0.0)
    
    fig = plt.figure(figsize=(15,15))

#     # Create figure for original image
#     ax1 = fig.add_subplot(1,8,1)
#     plt.title(img_file)
#     ax1.imshow(image)
    
#     # Create figure for grayscale image
#     ax2 = fig.add_subplot(1,8,2)
#     plt.title("RGB image after Color Masking")
#     ax2.imshow(img_out)
    
#     # Create figure for grayscale image
#     ax3 = fig.add_subplot(1,8,3)
#     plt.title('Grayscale')
#     ax3.imshow(gray_image, cmap='gray')
    
#     # Create figure for smoothed image
#     ax4 = fig.add_subplot(1,8,4)
#     plt.title('Gaussian Blur')
#     ax4.imshow(blur_gray_img, cmap='gray')
    
#     # Create figure for Canny Edge Detector
#     ax5 = fig.add_subplot(1,8,5)
#     plt.title('Canny Edge')
#     ax5.imshow(canny_edge_img, cmap='gray')
    
#     # Create figure for smoothed image
#     ax6 = fig.add_subplot(1,8,6)
#     plt.title('Region of Interest')
#     ax6.imshow(masked_edge_img, cmap='gray')
    
#      # Create figure for smoothed image
#     ax6 = fig.add_subplot(1,8,7)
#     plt.title('Hough Transform')
#     ax6.imshow(lines)
    
    # Create figure for smoothed image
    # Convert to HLS color space and mask yellow and white 
    hls_img = cv2.cvtColor(image, cv2.COLOR_RGB2HLS)
    
    #plt.title('Lines added to Original Image')
    plt.imshow(image)
    plt.savefig(output_path + 'original_image.jpg')
    
    if DEBUG:
        break
````
